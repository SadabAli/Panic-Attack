```python
from sklearn.compose import ColumnTransformer 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder , OneHotEncoder , OrdinalEncoder
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
X = df.drop(columns=['Panic_Score'])
y = df['Panic_Score']
```

```python
X.sample(2)
```

```python
y.sample(2)
```

```python
# Define the ColumnTransformer
transformer = ColumnTransformer(transformers=[
    # Impute missing values for Medical_History
    ('Medical_History_trf', SimpleImputer(strategy='most_frequent'), ['Medical_History']),
    
    # Encode binary columns (Yes/No)
    ('Label_Encoding', OrdinalEncoder(), ['Sweating', 'Shortness_of_Breath', 'Dizziness',
                                          'Chest_Pain', 'Trembling', 'Medication',
                                          'Smoking', 'Therapy']),
    
    # One-Hot Encode categorical columns
    ("OHE", OneHotEncoder(sparse_output=False, drop='first'), ['Trigger'])
], remainder='passthrough')

# Fit and transform the data
X_transformed = transformer.fit_transform(X)

# Optionally, convert back to a DataFrame
import pandas as pd
transformed_columns = (
    ['Medical_History'] +  # Output from SimpleImputer
    ['Sweating', 'Shortness_of_Breath', 'Dizziness', 'Chest_Pain', 'Trembling', 'Medication', 'Smoking', 'Therapy'] +  # Label Encoding
    transformer.named_transformers_['OHE'].get_feature_names_out(['Trigger']).tolist() +  # One-Hot Encoding
    list(X.columns.drop(['Medical_History', 'Sweating', 'Shortness_of_Breath', 'Dizziness',
                         'Chest_Pain', 'Trembling', 'Medication', 'Smoking', 'Therapy', 'Trigger']))  # Remaining columns
)
X = pd.DataFrame(X_transformed, columns=transformed_columns)
```

```python
X
```

```python
X['Medical_History'].value_counts()
```

```python
ohe = OneHotEncoder(sparse_output=True,drop='first')
```

```python
# Initialize the encoder
ohe = OneHotEncoder(sparse_output=False, drop='first')

# Fit and transform the column
encoded = ohe.fit_transform(X[['Medical_History']])

# Convert the output into a DataFrame
encoded_X = pd.DataFrame(encoded, columns=ohe.get_feature_names_out(['Medical_History']))

# Concatenate the encoded columns with the original DataFrame
X = pd.concat([X, encoded_X], axis=1)

# Drop the original column if no longer neededX.drop(columns=['Medical_History'], inplace=True)

print(X)
```

```python
print(X.shape)
X
```

```python
X.drop(columns=['Medical_History'],inplace=True)
```

```python
X_train , X_test  , y_train ,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score ,mean_absolute_error,mean_squared_error,r2_score
from sklearn.model_selection import cross_val_score
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

# Predict class labels
y_pred = model.predict(X_test)

print("mean_squared_error",mean_squared_error(y_test, y_pred))
print("mean_absolute_error",mean_absolute_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print("R-squared (R²):", r2)
```
